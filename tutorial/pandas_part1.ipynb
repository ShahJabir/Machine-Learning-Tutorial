## Pandas Basic Funtionalities


### Head Function in Pandas


```python
# head function
default_head_value = df.head()
three_head_value = df.head(3)
print("Dafault value in head\n")
print(default_head_value)
print("\n")
print("Three value in head\n")
print(three_head_value)
```

### Tail Function in Pandas


```python
# tail function
default_tail_value = df.tail()
three_tail_value = df.tail(3)
print("Dafault value in tail\n")
print(default_tail_value)
print("\n")
print("Three value in tail\n")
print(three_tail_value)
```

### Columns Function in Pandas to get columns of data


```python
# column
import numpy as np

df.columns
col = np.array(df.columns)
print("All columns of csv\n")
print(f"{col}\n")
print("Columns type of csv\n")
print(col.dtype)
```

### Index Function in Pandas to get index of data


```python
# index
ind = np.array(df.index)
print("All index of csv \n")
print(ind)
```

### Basic Function


```python
# info
print("Info function to get info of data src file \n")
print(df.info())
print("\n")

# sample
print("Get some default random sample of data \n")
print(df.sample())
print("\n")

# statistical values
print("Get statistical values of data \n")
print(df.describe())
print("\n")
```

## Transforming to df from inbuilt datas


### List Data


```python
my_list = [["Alice", 25], ["Bob", 30], ["Charlie", 28]]

list_df = pd.DataFrame(my_list, columns=["Name", "Age"], index=[1, 2, 3])

print("Get Data Frame from a List \n")
list_df
```

### Tuple Data


```python
my_tuple = (("Alice", 25), ("Bob", 30), ("Charlie", 28))

tuple_df = pd.DataFrame(my_tuple, columns=["Name", "Age"])
print("Get Data Frame from a Tuple \n")
tuple_df
```

### Dictionary Data


```python
my_dict = {"Name": ["Alice", "Bob", "Charlie"], "Age": [25, 29, 28]}

dict_df = pd.DataFrame(my_dict)
print("Get Data Frame from a Dictionary \n")
dict_df
```

### JSON Data


```python
json_data = [
    {"Name": "Alice", "Age": 25, "City": "New York"},
    {"Name": "Bob", "Age": 30, "City": "Paris"},
    {"Name": "Charlie", "Age": 28, "City": "Rajshahi"},
]

data_df = pd.DataFrame(json_data)
print("Get Data Frame from a JSON \n")
data_df
```
